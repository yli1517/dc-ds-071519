```python
df['date_new'] = pd.to_datetime(df.date)
```

```python
df['month'] = df.date_new.dt.to_period("M")
```

```python
df['amount_by_month'] = df.groupby(['month'])['amount'].transform('sum')
```

```python
df['amount_by_purpose'] = df.groupby(['purpose'])['amount'].transform('sum')
```

```python
df['amount_by_source'] = df.groupby(['source'])['amount'].transform('sum')
```
